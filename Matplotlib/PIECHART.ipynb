### Labels in pie chart

```python
slices=[120,60]
labels=["men","women"]
plt.pie(slices,labels=labels)  # (or)  plt.pie(slices,labels=["men","women"])
plt.show()
```

### EDGE COLORS

```python
plt.style.use('fivethirtyeight')
slices=[60,40]
labels=["MEN","WOMEN"]
plt.pie(slices,labels=labels,wedgeprops={'edgecolor':'black'})
plt.title("My Pie Chart")
plt.show()
```

### COLORS 

```python
slices=[20,50,60]
labels=["C","JAVA","PYTHON"]
colors=["#34deeb","blue","#eb3480"]
plt.pie(slices,labels=labels,colors=colors,wedgeprops={'edgecolor':'black'})
plt.show()
```

### EXPLODE

```python
plt.style.use('fivethirtyeight')
slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python','other']
explode=[0,0,0,0.1,0]
plt.pie(slices,labels=labels,explode=explode)
plt.show()
```

```python
plt.style.use('fivethirtyeight')
slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python','other']
explode=[0,0,0,0.5,0]
plt.pie(slices,labels=labels,explode=explode)
plt.show()
```

### SHADOW

```python
plt.style.use('fivethirtyeight')
slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python','other']
explode=[0,0,0,0.1,0]
plt.pie(slices,labels=labels,explode=explode,shadow=True)
plt.show()
```

### PERCENTAGE

```python
plt.style.use('fivethirtyeight')
slices = [59219,36430,47544,55466,35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python','other']
explode=[0,0,0,0.1,0]
plt.pie(slices,labels=labels,explode=explode,autopct='%1.1f%%')
plt.title('My PieChart')
plt.show()
```

```python
plt.style.use('fivethirtyeight')
slices = [59219,36430,47544,55466,35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python','other']
explode=[0,0,0,0.1,0]
plt.pie(slices,labels=labels,explode=explode,autopct='%1.1f%%')
plt.title('My PieChart')
plt.tight_layout()
plt.show()
```
