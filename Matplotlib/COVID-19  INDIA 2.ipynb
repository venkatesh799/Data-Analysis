```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("covid_19_india 26 mar.csv")
data
```

```python
data['Date']=data['Date'].map(pd.to_datetime)
```

```python
data['State/UnionTerritory'].nunique()
```

```python
plt.figure(figsize=(19,12))
plt.style.use('seaborn')
data[data['Date'] == '2020-03-26'].groupby('State/UnionTerritory').ConfirmedIndianNational.sum().sort_values(ascending=False).plot(kind='bar',color="#fcdb03",width=0.8)
plt.xticks(size=18)
plt.title("Active Cases",size=20)
plt.xlabel("State/UnionTerritory",size=18)
plt.ylabel("Cases",size=18)
```
