```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime
import seaborn as sns
```

```python
data=pd.read_csv("covid_19_india.csv")
data.tail()
```

```python
plt.style.use('fivethirtyeight')
indian=data.ConfirmedIndianNational.sum()
foreign=data.ConfirmedForeignNational.sum()
plt.pie([indian,foreign],labels=['Indian','Foreign Nationals'],autopct='%1.1f%%')
plt.show()
```

```python
data['TotalCases']=data['ConfirmedIndianNational']+data['ConfirmedForeignNational']
data.head()
```

```python
plt.figure(figsize=(19,12))
colors=data['TotalCases']
plt.style.use('fivethirtyeight')
plt.bar(data['State/UnionTerritory'],data['TotalCases'],edgecolor='black')
plt.xlabel("No.of Cases",fontsize=20)
plt.ylabel("State/UnionTerritory",fontsize=20)
plt.xticks(rotation=90,size=20)
plt.show()
```

```python
plt.figure(figsize=(20,12))
plt.style.use('seaborn')
confirmed_cases=data['TotalCases']
plt.plot(confirmed_cases)
plt.xlabel("Day Since 30/01/2020",fontsize=20)
plt.ylabel("No Of Cases",fontsize=18)
plt.xticks(size=20)
plt.yticks(size=20)
plt.title("Corona Viruses OverTime",size=30)
plt.show()
```

```python
plt.figure(figsize=(20,12))
plt.style.use("seaborn")
deaths=data.ConfirmedIndianNational
plt.plot(deaths)
plt.xlabel("Day Since 30/01/2020",size=20)
plt.ylabel("No Of Cases",size=20)
plt.xticks(size=20)
plt.yticks(size=20)
plt.title(".Confirmed Indian National",size=20)
plt.show()
```

```python
plt.figure(figsize=(20,12))
plt.style.use("seaborn")
foreign=data.ConfirmedForeignNational
plt.plot(foreign)
plt.xlabel("Day Since 30/01/2020",size=20)
plt.ylabel("No Of Cases",size=20)
plt.title("Confirmed Foreign National",size=20)
plt.xticks(size=20)
plt.yticks(size=20)
plt.show()
```

```python
plt.style.use("seaborn")
plt.figure(figsize=(20,12))
plt.bar(data['State/UnionTerritory'],data['Cured'])
plt.xlabel('State/UnionTerritory',size=20)
plt.ylabel("No Of Cured",size=20)
plt.title("Cured StateWise/UnionTerritory",size=20)
plt.xticks(rotation=90,size=20)
plt.yticks(size=20)
plt.show()
```

```python
plt.style.use("seaborn")
plt.figure(figsize=(20,12))
plt.bar(data['State/UnionTerritory'],data['Deaths'])
plt.xlabel('State/UnionTerritory',size=20)
plt.ylabel("No Of Cured",size=20)
plt.title("Deaths StateWise/UnionTerritory",size=20)
plt.xticks(rotation=90,size=20)
plt.yticks(size=20)
plt.show()
```

```python
plt.figure(figsize=(20,12))
plt.style.use("seaborn")
plt.plot(data['Cured'])
plt.plot(data['Deaths'])
plt.show()
```
