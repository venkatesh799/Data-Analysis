```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Sales.csv")
```

```python
data.head()
```

```python
data.columns
```

### Data Wrangling

```python
data.drop(columns = ["Column1"], inplace = True)
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
data.shape
```

```python
data.dropna(inplace = True)
```

```python
data.isnull().sum()
```

```python
# converting Date to data time format
data["Date"] = pd.to_datetime(data["Date"])
data.head()
```

```python
# Creating the dayofthe week
data["Dayofweek"] = data["Date"].dt.dayofweek
data.head()
```

```python
# Name of the Weekday creation
days = {0:"Monday", 1:"Tuesday", 2:"Wednesday", 3:"Thursday", 4:"Friday", 5:"Saturday", 6:"Sunday"}
data["Weekday"] = data["Date"].dt.weekday.map(days)
data["Year"] = data["Date"].dt.year
data.head()
```

```python
data["Profit"] = data["Revenue"] - data["Cost"]
data.head()
```

### Analysis

```python
# Product Category 
product_category = data.groupby(["Product Category", "Customer Gender"])["Profit"].sum().reset_index().sort_values(by = "Profit", ascending = False)
product_category
```

```python
sns.barplot(x = "Product Category", y = "Profit", hue = "Customer Gender", data = product_category)
plt.title("Product Category Profits")
```

```python
data["Sub Category"].nunique()
```

```python
sub_category = data.groupby(["Sub Category", "Customer Gender"])["Profit"].sum().reset_index().sort_values(by = "Profit", ascending = False).head(20).reset_index(drop = True)
sub_category
```

```python
fig = plt.figure(figsize=(10, 5))
sns.barplot(x = "Profit", y = "Sub Category", hue = "Customer Gender", data = sub_category)
plt.title("Sub Category wise Profit with Gender")
```

```python
country = data.groupby(["Country"])["Profit"].sum().reset_index()
sns.barplot(x = "Country", y = "Profit", data = country)
plt.title("Country Wise Profits")
plt.ticklabel_format(style='plain', axis='y')
```

```python
country_gender = data.groupby(["Country", "Customer Gender"])["Profit"].sum().reset_index()
sns.barplot(x = "Country", y = "Profit", hue = "Customer Gender", data = country_gender)
plt.title("Country Wise Profit with Gender")
plt.ticklabel_format(style='plain', axis='y')
```

```python
date_year = data.groupby(["Year", "Month"])["Profit"].sum().reset_index()
months = {"January":1, "February": 2, "March":3, "April":4, "May":5, "June":6, "July":7, "August":8, "September":9, "October":10, "November":11, "December":12}
date_year["Month_Number"] = date_year["Month"].map(months)
```

```python
date_year = date_year.sort_values(by = "Month_Number").reset_index()
date_year
```

```python
plt.figure(figsize=(10,6))
plt.xticks(rotation = 90)
sns.lineplot(x = date_year["Month"], y = "Profit", data = date_year[date_year["Year"] == 2015], label = "2015")
sns.lineplot(x = date_year["Month"], y = "Profit", data = date_year[date_year["Year"] == 2016], label = "2016")
sns.set_theme(style="whitegrid")
plt.title("Profit Comparison: 2015 vs 2016")
plt.show()
```
