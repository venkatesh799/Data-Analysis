```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data=pd.read_csv("patient.csv")
data.head()
```

```python
data['Age']=2020.0-data['birth_year']
```

```python
data1=data.copy()
data1.head()
```

```python
plt.figure(figsize=(20,10))
x=list(data1.region.value_counts().sort_values(ascending=False).index)
y=list(data1.region.value_counts().sort_values(ascending=False))
sns.barplot(y,x)
plt.title("RegionWise Confirmed Cases",fontsize=20)
plt.show()
```

```python
plt.figure(figsize=(12,7))
sns.countplot(data1['state'])
plt.yscale('log')
plt.show()
```

```python
plt.figure(figsize=(12,7))
sns.countplot(data1['sex'],hue=data1.state)
plt.show()
```
